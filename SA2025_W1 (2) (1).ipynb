### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Cars.csv')

# Calculate the average mpg per year per origin
avg_mpg_yearly_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Design the multi-line plot
plt.figure(figsize=(12, 6)) # Set the figure size for better readability
sns.lineplot(data=avg_mpg_yearly_origin, x='model_year', y='mpg', hue='origin', marker='o')

# Add title and labels
plt.title('Evolution of Average MPG Over Years by Car Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(avg_mpg_yearly_origin['model_year'].unique()) # Ensure all years are marked on x-axis
plt.grid(True, linestyle='--', alpha=0.7) # Add a grid for better readability
plt.legend(title='Origin') # Add a legend for the origins

# Display the plot
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Cars.csv')

# Create the Seaborn scatterplot
plt.figure(figsize=(10, 7)) # Adjust figure size for better visualization
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 400), # Adjust the range of sizes for the dots
    hue_order=['japan', 'europe', 'usa'],
    alpha=0.7 # Add some transparency
)

# Add meaningful plot titles and axis titles
plt.title('Horsepower vs. Weight of Cars, Colored by Origin and Sized by MPG')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin / MPG', bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside the plot
plt.grid(True, linestyle='--', alpha=0.7) # Add a grid for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping

# Display the plot
plt.show()
```
